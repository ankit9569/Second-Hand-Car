##### Also Plot the countplot based on the Fuel Type

```python
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
sb.countplot(x=df['Fuel Type'],data=df,hue='Fuel Type')
plt.xlabel("Fuel Type",color='r',size=15)
plt.ylabel("No. Of Car",color='red',size=15)
plt.show()
```

```python
df.head()
```

### Display the  name and Price of  Car which have Mileage>=70000 and Body Color is Black

```python
car_filtered=df[(df['Mileage']>70000)&(df['Body Color']=='Black')&(df['Company Name'].isin(['Maruti Suzuki']))][['Car Name','Price','Company Name']]
car_filtered
```

```python
sb.countplot(x='Car Name',data=car_filtered,hue='Car Name',)
plt.xlabel('Car Name',color='r',size=15)
plt.ylabel('No. of Car(Milege>70000)',color='r',size=15)
plt.show()
```

### Display the number of car based on the Owner Type Company Wise

```python
res_df=df.groupby(by='Company Name')[['Owner Type']].size().reset_index(name='No of Car')
res_df.set_index('Company Name')
```
